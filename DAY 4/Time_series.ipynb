#### Conversion to DateTime

```python
df['DateTime']=pd.to_datetime(df['DateTime'])
df['DateTime']
```

#### Extracted individual components

```python
df['Year']=df['DateTime'].dt.year
df['Month']=df['DateTime'].dt.month
df['Time']=df['DateTime'].dt.time
df['Hour']=df['DateTime'].dt.hour
df['Date']=df['DateTime'].dt.date
```

####  Basic Date Operations

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
df[df['Year']>2022]
```

####  Set up Index & Head

```python
df=df.set_index('DateTime')
df.head()
```

```python
df=df.drop(['Date'],axis=1)
```

#### Resampled the data

```python
daily_stats=df['Consumption'].resample('M').agg(['sum','mean','count'])
print(daily_stats)
```

#### Implemented Aggregated Functions

```python
summary_stats_consumption=df['Consumption'].agg(['sum','min','max'])
summary_stats_production=df['Production'].agg(['mean','std'])
print(summary_stats_consumption)
print(summary_stats_production)
```

```python
df.reset_index(inplace=True)
df.head()
```

#### Timezone Handling

```python
df['DateTime']=df['DateTime'].dt.tz_localize('UTC')
df['DateTime']
```

```python
df['DateTime']=df['DateTime'].dt.tz_convert('Asia/Kolkata')
df['DateTime']
```

#### Time Series Functions

```python
df['Rolling_consumption_mean']=df['Consumption'].rolling(window=5).mean()
df['Rolling_consumption_mean']
```

```python
df['Prev_year']=df['Year'].shift(1)
df['Prev_year']
```
